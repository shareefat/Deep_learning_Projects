```python
import os

dataset_path = "/content/drive/MyDrive/Classroom/PGDDM 44 python/valid"
names = os.listdir(dataset_path)
```

```python
names
```

```python
data = []
labels = []

for category in names:
  folder_path = os.path.join(dataset_path, category)
  for file in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file)
    data.append(file_path)
    labels.append(category)
```

```python
print(data[:5], labels[:5])
```

```python
label_map = {}
for id, category in enumerate(names):
  label_map[category] = id
```

```python
from tensorflow.keras.preprocessing.image import img_to_array,load_img

def preprocess_image(data):
  img = load_img(data,target_size = (150,150))
  img_array = img_to_array(img)
  img_array = img_array/255.0
  return img_array
```

```python
import numpy as np
```

```python
y = []
for i in labels:
  y.append(label_map[i])
y = np.array(y)
y[:5]
```

```python
data
```

```python
x = []
for img in data:
  x.append(preprocess_image(img))
x = np.array(x)
x[:5]
```

```python
x.shape
```

```python
import matplotlib.pyplot as plt

plt.imshow(x[1])
plt.show
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, AveragePooling2D

model = tf.keras.Sequential([
            Conv2D(32,(3,3), activation = 'relu', input_shape = (150,150,3)),
            MaxPooling2D((2,2)),
            Conv2D(32,(3,3), activation = 'relu', input_shape = (150,150,3)),
            MaxPooling2D((2,2)),
            Conv2D(32,(3,3), activation = 'relu', input_shape = (150,150,3)),
            AveragePooling2D((2,2)),
            Flatten(),
            Dense(128, activation='relu'),
            Dense(len(names),activation = "softmax")])
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics =['accuracy'])
```

```python
model.fit(x ,y , epochs=10)
```
